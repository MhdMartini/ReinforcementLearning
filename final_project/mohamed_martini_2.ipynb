## Overview

### Small Env

For this part a discrete 2D ($N \times M$) environment was provided, with dimensions $ N = M = 15$. The provided environment did not explicitly provide a terminal state, so I assumed it is the unique cell colored by "goal color", as seen in the figures below. The reward profile provided is also consistent with this being the terminal state. If no terminal state was provided to the learner, it will just learn to constantly move back and forth into the green cell. 


<h5 align="center">
    Figure 1 - The agent (yellow) needs to find its way to the goal state (green) while avoiding the red cells. Grey cells are walls that the agent cannot go through.  
</h5>


The challenge in this environment lies in the little tunnles created by the vertical walls. The probability of visiting cells between walls is higher that other cells, so unless the learning rate is carefully tuned, the agent can easily get stuck in the corners or between walls. 

The state vector $\textbf{s}$ is composed of two elements: the row and column of the agent 

$$
\textbf{s}_t = (row_{a_t} \in \mathcal{R}^N , col_{a_t} \in \mathcal{R}^M); \in \mathcal{R}^{15 \times 15}
$$

The agent can move in any of the four directions: 

$$
A = \{\rightarrow, \uparrow, \downarrow, \leftarrow\}
$$

The reward function is:

$$
R = 
\begin{cases}
    -0.1 & \text{each time step} \\
    -5.0 & \text{touching fire} \\
    +0 & \text{terminal state}
\end{cases}
$$

### Large Env

This environment is a scaled up version of the small environment ($30 \times 30$), with the main difference being the stochasticity in the blue cells (ice). When the agent takes an action on a blue cell, there is a 30% chance of it getting a random action instead, which could result in stepping on the red cells and getting a negative reward. When the agent spawns in the top left as shown, it should learn to avoid the stochastic region all together and find its way to the terminal state through the bottom passage in the map.


<h5 align="center">
    Figure 2 - Large Provided Grid World. Blue cells are stochastic - the agent has 70% chance of taking the action it intends, and 30% chance of ending up with a random action, and possibly stepping on fire. 
</h5>


The state vector and reward function are the same as those for the small environment.
